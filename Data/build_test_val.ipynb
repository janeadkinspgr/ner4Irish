```python
def read_conll_file(file_path):
    data = []
    current_sentence = []
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            line = line.strip()
            if line.startswith('-DOCSTART-'):
                continue
            if line:
                parts = line.split()
                word = parts[0]
                ner_label = parts[-1]
                current_sentence.append((word, ner_label))
            else:
                if current_sentence:
                    data.append(current_sentence)
                    current_sentence = []
    if current_sentence:
        data.append(current_sentence)
    return data
```

```python
IUDT_val = read_conll_file('/content/IUDT_test.conll')[:40]
IUDT_test = read_conll_file('/content/IUDT_test.conll')[40:]
print(len(IUDT_val))
print(len(IUDT_test))
print(IUDT_val)
print(IUDT_test)
```

```python
dail_val = read_conll_file('/content/additional_Dail.conll')[:60]
dail_test = read_conll_file('/content/additional_Dail.conll')[60:]
print(len(dail_val))
print(len(dail_test))
print(dail_val)
print(dail_test)
```

```python
#building the sets

val = IUDT_val + dail_val
print(len(val))
print(val)
```

```python
def write_conll_file(sentences, output_file):
    with open(output_file, 'w', encoding='utf-8') as f:
        for sentence in sentences:
            for word, tag in sentence:
                f.write(word + '\t' + tag + '\n')
            f.write('\n')
```

```python
write_conll_file(val, 'NER_Irish_validation.conll')
```

```python
write_conll_file(test, 'NER_Irish_test.conll')
```
