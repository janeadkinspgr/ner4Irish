```python
def extract_sentences_with_named_entities(conllu_file):
    sentences = []
    current_sentence = []

    with open(conllu_file, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line.startswith('# sent_id'):
                current_sentence = []
            if line == '' and current_sentence:
                sentences.append(current_sentence)
                current_sentence = []
            current_sentence.append(line)

    # Check if each sentence contains "NamedEntity=Yes" in any of its lines
    sentences_with_named_entities = [sentence for sentence in sentences if any("NamedEntity=Yes" in line for line in sentence)]

    return sentences_with_named_entities

conllu_file = 'ga_idt-ud-test.conllu'
result = extract_sentences_with_named_entities(conllu_file)

for sentence in result:
    text_line = None
    for line in sentence:
        if line.startswith('# text ='):
            text_line = line
            break

    if text_line:
        text = text_line.replace('# text = ', '')
        #print(text)

print(len(result))

output_file = 'iudt_test.txt'
with open(output_file, 'w', encoding='utf-8') as output:
    for sentence in result:
        text_line = None
        for line in sentence:
            if line.startswith('# text ='):
                text_line = line
                break

        if text_line:
            text = text_line.replace('# text = ', '')
            output.write(text + '\n')

print("Selected sentences have been written to", output_file)
```
