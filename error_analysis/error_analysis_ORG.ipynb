```python
def read_conll_file(file_path):
    data = []
    current_sentence = []
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            line = line.strip()
            if line.startswith('-DOCSTART-'):
                continue
            if line:
                parts = line.split()
                word = parts[0]
                gold_label = parts[-2]
                pred_label = parts[-1]
                current_sentence.append((word, gold_label, pred_label))
            else:
                if current_sentence:
                    data.append(current_sentence)
                    current_sentence = []
    if current_sentence:
        data.append(current_sentence)
    return data
```

```python
def fill_dictionaries(fp, fn):
  #dict
  dict_name = {'fp': [], 'fn': []}
  #fp
  for k,v in fp.items():
    for vals in v:
      dict_name['fp'].append(vals)
  #fn
  for k,v in fn.items():
    for vals in v:
      dict_name['fn'].append(vals)
  return dict_name
```
